<h1>Loyalty</h1>

```python
person=pd.read_csv('person_outer.csv')
qpoll=pd.read_csv('qpoll_outer_자연어변환.csv')
```

```python
full=pd.merge(person,qpoll, on='고유번호', how='outer')
```

```python
full['loyalty'] = full.notna().sum(axis=1)
```

```python
full=full.sort_values(by='loyalty',ascending=False)
full = full.reset_index(drop=True)
```

```python
full.to_csv('full_outer_loyalty.csv')
```
